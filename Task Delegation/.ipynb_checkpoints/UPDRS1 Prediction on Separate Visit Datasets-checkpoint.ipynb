```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def generate_point_plot(file_path, image_name, metric_name):
    # Load the data from the Excel file
    df = pd.read_excel(file_path)

    # Verify the structure of the DataFrame
    print(df.head())

    # Melt the DataFrame to long format
    melted_df = df.melt(id_vars=['Dimensionality Reduction', 'Dataset'], 
                        value_vars=['Linear Regression', 'SVR', 'Random Forest', 'XGBoost'],
                        var_name='Model', value_name=metric_name)

    # Assign a color based on the dimensionality reduction technique
    def assign_color(row):
        if 'NMF' in row['Dimensionality Reduction'] and 'PCA' in row['Dimensionality Reduction']:
            return 'PCA + NMF'
        elif 'NMF' in row['Dimensionality Reduction']:
            return 'NMF'
        elif 'PCA' in row['Dimensionality Reduction']:
            return 'PCA'
        return 'Other'

    melted_df['Color Group'] = melted_df.apply(assign_color, axis=1)

    # Define a color palette
    palette = {
        'NMF': '#1f77b4',
        'PCA': '#ff7f0e',
        'PCA + NMF': '#2ca02c'
    }

    # Create the point plot
    plt.figure(figsize=(14, 8))
    sns.scatterplot(data=melted_df, x='Dataset', y=metric_name, hue='Color Group', palette=palette, s=100)

    plt.title(f'{metric_name} Scores of Different Models Across Datasets')
    plt.xlabel('Dataset')
    plt.ylabel(f'{metric_name} Score')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Dimensionality Reduction', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

    # Save the point plot as an image file
    plt.savefig(image_name)
    plt.show()



```

```python
import warnings

warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn.linear_model._base')
warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn.linear_model._base')
warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn.decomposition._nmf')

def run_experiments(dataframes, target_column_name):

    results = []

    for name, df in dataframes:
        # PCA
        best_model_pca, best_params_pca, best_score_pca, mse_pca, r2_pca, rmse_pca, mae_pca, rank_order_corr_pca, adjusted_r2_pca = complete_workflow_pca(df, target_column_name)
        results.append({
            'Dataset': name,
            'Method': 'PCA',
            'Best_Model': best_model_pca,
            'Best_Params': best_params_pca,
            'Best_Score': best_score_pca,
            'MSE': mse_pca,
            'R2': r2_pca,
            'RMSE': rmse_pca,
            'MAE': mae_pca,
            'Rank-Order Corr': rank_order_corr_pca,
            'Adjusted R': adjusted_r2_pca
        })

        # NMF
        best_model_nmf, best_params_nmf, best_score_nmf, mse_nmf, r2_nmf, rmse_nmf, mae_nmf, rank_order_corr_nmf, adjusted_r2_nmf = complete_workflow_nmf(df, target_column_name)
        results.append({
            'Dataset': name,
            'Method': 'NMF',
            'Best_Model': best_model_nmf,
            'Best_Params': best_params_nmf,
            'Best_Score': best_score_nmf,
            'MSE': mse_nmf,
            'R2': r2_nmf,
            'RMSE': rmse_nmf,
            'MAE': mae_nmf,
            'Rank-Order Corr': rank_order_corr_nmf,
            'Adjusted R': adjusted_r2_nmf
        })

        # PCA+NMF Multiple Combination
        best_model_multi_combine, best_params_multi_combine, best_score_multi_combine, mse_multi_combine, r2_multi_combine, rmse_multi_combine, mae_multi_combine, rank_order_corr_multi_combine, adjusted_r2_multi_combine = complete_workflow_multiple_combination(df, target_column_name)
        results.append({
            'Dataset': name,
            'Method': 'PCA+NMF-Multiple Combo',
            'Best_Model': best_model_multi_combine,
            'Best_Params': best_params_multi_combine,
            'Best_Score': best_score_multi_combine,
            'MSE': mse_multi_combine,
            'R2': r2_multi_combine,
            'RMSE': rmse_multi_combine,
            'MAE': mae_multi_combine, 
            'Rank-Order Corr': rank_order_corr_multi_combine,
            'Adjusted R': adjusted_r2_multi_combine
        })

    results_df = pd.DataFrame(results)

    results_df.to_csv('results_summary.csv', index=False)
    
    return results_df
```

```python
cross_sectional_dataframes = [
    ('BL', cross_bl_dataframe), 
    ('V04', cross_v04_dataframe), 
    ('V06', cross_v06_dataframe), 
    ('V08', cross_v08_dataframe), 
    ('V10', cross_v10_dataframe), 
    ('V12', cross_v12_dataframe)
]

results_df = run_experiments(cross_sectional_dataframes, 'UPDRS')
results_summary_df = results_df[['Dataset', 'Method', 'MSE', 'R2', 'RMSE', 'MAE', 'Rank-Order Corr', 'Adjusted R']]
print(results_summary_df)
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

def plot_results(results_df, arrangement_type):
    # Set vibrant color palette
    sns.set_palette("bright")
    
    # Define the directory to save plots
    save_dir = 'D:\\msc_thesis_plots_results'
    os.makedirs(save_dir, exist_ok=True)
    
    # List of metrics to plot
    metrics = ['MSE', 'R2', 'RMSE', 'MAE', 'Rank-Order Corr', 'Adjusted R']

    for metric in metrics:
        plt.figure(figsize=(12, 6))
        ax = sns.barplot(x='Dataset', y=metric, hue='Method', data=results_df)
        
        # Add value annotations on top of the bars
        for p in ax.patches:
            height = p.get_height()
            ax.annotate(f'{height:.2f}', 
                        (p.get_x() + p.get_width() / 2., height), 
                        ha='center', va='center', 
                        xytext=(0, 9), 
                        textcoords='offset points')

        plt.xlabel('Dataset')
        plt.ylabel(metric)
        plt.title(f'{metric} Comparison Across Methods')
        plt.legend(title='Method')
        plt.xticks(rotation=45)
        plt.tight_layout()
        
        plot_name = f'{arrangement_type}_{metric}.png'
        plt.savefig(os.path.join(save_dir, plot_name), dpi=300)
        
        plt.show()
```

```python
def drop_updrs_2_3_4_from_wide_format(df):
    updrs2_columns = ["NP2SPCH", "NP2SALV", "NP2SWAL", "NP2EAT", "NP2DRES", "NP2HYGN", "NP2HWRT",
                  "NP2HOBB", "NP2TURN", "NP2TRMR", "NP2RISE", "NP2WALK", "NP2FREZ"]
    
# "['CMEDTM', 'EXAMTM', 'PN3RIGRL', 'DYSKPRES', 'DYSKIRAT', 'ANNUAL_TIME_BTW_DOSE_NUPDRS', 'ON_OFF_DOSE', 'PD_MED_USE']
    updrs3_columns = ["NP3SPCH", "NP3FACXP", "NP3RIGN", "NP3RIGRU", "NP3RIGLU", "NP3RIGLL",
                  "NP3FTAPR", "NP3FTAPL", "NP3HMOVR", "NP3HMOVL", "NP3PRSPR", "NP3PRSPL", "NP3TTAPR", "NP3TTAPL", "NP3LGAGR",
                  "NP3LGAGL", "NP3RISNG", "NP3GAIT", "NP3FRZGT", "NP3PSTBL", "NP3POSTR", "NP3BRADY", "NP3PTRMR", "NP3PTRML",
                  "NP3KTRMR", "NP3KTRML", "NP3RTARU", "NP3RTALU", "NP3RTARL", "NP3RTALL", "NP3RTALJ", "NP3RTCON",
                   "NHY"]

#     updrs4_columns = ["NP4WDYSK", "NP4DYSKI", "NP4OFF", "NP4FLCTI", "NP4FLCTX", "NP4DYSTN"]
    df = df.drop(columns=updrs2_columns + updrs3_columns)
    
    return df

def add_updrs1_column_in_wide_format(df):
    updrs1_columns = ['NP1COG', 'NP1HALL', 'NP1DPRS', 'NP1ANXS', 'NP1APAT', 'NP1DDS',
                      'NP1SLPN', 'NP1SLPD', 'NP1PAIN', 'NP1URIN', 'NP1CNST', 'NP1LTHD', 'NP1FATG']
    
    updrs1_multiindex = pd.MultiIndex.from_product([['UPDRS1'], df.columns.levels[1]], names=df.columns.names)

    updrs1_df = pd.DataFrame(index=df.index, columns=updrs1_multiindex)
    
    for visit in df.columns.levels[1]:
        updrs1_df[('UPDRS1', visit)] = df.loc[:, (updrs1_columns, visit)].sum(axis=1)
    
    df = df.drop(columns=updrs1_columns)
    
    df = pd.concat([df, updrs1_df], axis=1)
    
    return df

def modify_dataframe(df, columns_to_drop, new_column_name):

    updrs_columns = {
        'UPDRS1': ['NP1COG', 'NP1HALL', 'NP1DPRS', 'NP1ANXS', 'NP1APAT', 'NP1DDS', 'NP1SLPN', 'NP1SLPD', 'NP1PAIN', 'NP1URIN', 'NP1CNST', 'NP1LTHD', 'NP1FATG'],
        'UPDRS2': ['NP2SPCH', 'NP2SALV', 'NP2SWAL', 'NP2EAT', 'NP2DRES', 'NP2HYGN', 'NP2HWRT', 'NP2HOBB', 'NP2TURN', 'NP2TRMR', 'NP2RISE', 'NP2WALK', 'NP2FREZ'],
        'UPDRS3': ['NP3SPCH', 'NP3FACXP', 'NP3RIGN', 'NP3RIGRU', 'NP3RIGLU', 'NP3RIGLL', 'NP3FTAPR', 'NP3FTAPL', 'NP3HMOVR', 'NP3HMOVL', 'NP3PRSPR', 'NP3PRSPL', 'NP3TTAPR', 'NP3TTAPL', 'NP3LGAGR', 'NP3LGAGL', 'NP3RISNG', 'NP3GAIT', 'NP3FRZGT', 'NP3PSTBL', 'NP3POSTR', 'NP3BRADY', 'NP3PTRMR', 'NP3PTRML', 'NP3KTRMR', 'NP3KTRML', 'NP3RTARU', 'NP3RTALU', 'NP3RTARL', 'NP3RTALL', 'NP3RTALJ', 'NP3RTCON', 'NHY']
        #'UPDRS4': ['NP4WDYSK', 'NP4DYSKI', 'NP4OFF', 'NP4FLCTI', 'NP4FLCTX', 'NP4DYSTN']
    }
    
    drop_column_names = []
    for drop_target in columns_to_drop:
            drop_column_names.extend(updrs_columns[drop_target])
    
    df = df.drop(columns=drop_column_names)
    
    if new_column_name in updrs_columns:
        new_column_set = updrs_columns[new_column_name]
        new_column_multiindex = pd.MultiIndex.from_product([[new_column_name], df.columns.levels[1]], names=df.columns.names)
        new_column_df = pd.DataFrame(index=df.index, columns=new_column_multiindex)
        
        for visit in df.columns.levels[1]:
            new_column_df[(new_column_name, visit)] = df.loc[:, (new_column_set, visit)].sum(axis=1)
        
        df = df.drop(columns=new_column_set)
        df = pd.concat([df, new_column_df], axis=1)
    
    return df



def extract_and_drop_target_from_wide_format(df, target_column_name, visit):
    # Extract the target column
    target_column = df[(target_column_name, visit)]
    
    # Drop the target column from the main dataframe
    df = df.drop(columns=[(target_column_name, visit)])
    
    return df, target_column

```

```python
def create_wide_format(df, columns_to_drop, target_column_name):
    wide_format_dataframe = modify_dataframe(df, columns_to_drop, target_column_name)

    wide_bl_columns = [col for col in wide_format_dataframe.columns if 'BL' in col]
    wide_bl_v04_columns = [col for col in wide_format_dataframe.columns if 'BL' in col or 'V04' in col]
    wide_bl_v04_v06_columns = [col for col in wide_format_dataframe.columns if 'BL' in col or 'V04' in col or 'V06' in col]
    wide_bl_v04_v06_v08_columns = [col for col in wide_format_dataframe.columns if 'BL' in col or 'V04' in col or 'V06' in col or 'V08' in col]
    wide_bl_v04_v06_v08_v10_columns = [col for col in wide_format_dataframe.columns if 'BL' in col or 'V04' in col or 'V06' in col or 'V08' in col  or 'V10' in col]
    wide_bl_v04_v06_v08_v10_v12_columns = [col for col in wide_format_dataframe.columns if 'BL' in col or 'V04' in col or 'V06' in col or 'V08' in col  or 'V10' in col or 'V12' in col]

    wide_bl_dataframe = wide_format_dataframe[wide_bl_columns]
    wide_bl_v04_dataframe = wide_format_dataframe[wide_bl_v04_columns]
    wide_bl_v04_v06_dataframe = wide_format_dataframe[wide_bl_v04_v06_columns]
    wide_bl_v04_v06_v08_dataframe = wide_format_dataframe[wide_bl_v04_v06_v08_columns]
    wide_bl_v04_v06_v08_v10_dataframe = wide_format_dataframe[wide_bl_v04_v06_v08_v10_columns]
    wide_bl_v04_v06_v08_v10_v12_dataframe = wide_format_dataframe[wide_bl_v04_v06_v08_v10_v12_columns]
    
    wide_bl_dataframe, wide_bl_target_column = extract_and_drop_target_from_wide_format(wide_bl_dataframe, target_column_name, 'BL')

    wide_bl_v04_dataframe, wide_bl_v04_target_column = extract_and_drop_target_from_wide_format(wide_bl_v04_dataframe, target_column_name, 'V04')

    wide_bl_v04_v06_dataframe, wide_bl_v04_v06_target_column = extract_and_drop_target_from_wide_format(wide_bl_v04_v06_dataframe, target_column_name, 'V06')

    wide_bl_v04_v06_v08_dataframe, wide_bl_v04_v06_v08_target_column = extract_and_drop_target_from_wide_format(wide_bl_v04_v06_v08_dataframe, target_column_name, 'V08')

    wide_bl_v04_v06_v08_v10_dataframe, wide_bl_v04_v06_v08_v10_target_column = extract_and_drop_target_from_wide_format(wide_bl_v04_v06_v08_v10_dataframe, target_column_name, 'V10')

    wide_bl_v04_v06_v08_v10_v12_dataframe, wide_bl_v04__v06_v08_v10_v12_target_column = extract_and_drop_target_from_wide_format(wide_bl_v04_v06_v08_v10_v12_dataframe, target_column_name, 'V12')
    
    wide_dataframes = [
        ('BL', wide_bl_dataframe, wide_bl_target_column), 
        ('BL_V04', wide_bl_v04_dataframe, wide_bl_v04_target_column), 
        ('BL_V04_V06', wide_bl_v04_v06_dataframe, wide_bl_v04_v06_target_column), 
        ('BL_V04_V06_V08', wide_bl_v04_v06_v08_dataframe, wide_bl_v04_v06_v08_target_column), 
        ('BL_V04_V06_V08_V10', wide_bl_v04_v06_v08_v10_dataframe, wide_bl_v04_v06_v08_v10_target_column), 
        ('BL_V04_V06_V08_V10_V12', wide_bl_v04_v06_v08_v10_v12_dataframe, wide_bl_v04__v06_v08_v10_v12_target_column)
        ]
    
    return wide_dataframes
```

```python
def run_wide_format_experiments(dataframes, model_name='LR'):

    results = []

    for name, df, target_column in dataframes:
        # PCA
        best_model_pca, best_params_pca, best_score_pca, mse_pca, r2_pca, rmse_pca, mae_pca, rank_order_corr_pca, adjusted_r2_pca = run_pca(df, target_column, model_name)
        results.append({
            'Dataset': name,
            'Method': 'PCA',
            'Best_Model': best_model_pca,
            'Best_Params': best_params_pca,
            'Best_Score': best_score_pca,
            'MSE': mse_pca,
            'R2': r2_pca,
            'RMSE': rmse_pca,
            'MAE': mae_pca,
            'Rank-Order Corr': rank_order_corr_pca,
            'Adjusted R': adjusted_r2_pca
        })

        # NMF
        best_model_nmf, best_params_nmf, best_score_nmf, mse_nmf, r2_nmf, rmse_nmf, mae_nmf, rank_order_corr_nmf, adjusted_r2_nmf = run_nmf(df, target_column, model_name)
        results.append({
            'Dataset': name,
            'Method': 'NMF',
            'Best_Model': best_model_nmf,
            'Best_Params': best_params_nmf,
            'Best_Score': best_score_nmf,
            'MSE': mse_nmf,
            'R2': r2_nmf,
            'RMSE': rmse_nmf,
            'MAE': mae_nmf,
            'Rank-Order Corr': rank_order_corr_nmf,
            'Adjusted R': adjusted_r2_nmf
        })

        # PCA+NMF Multiple Combination
        best_model_multi_combine, best_params_multi_combine, best_score_multi_combine, mse_multi_combine, r2_multi_combine, rmse_multi_combine, mae_multi_combine, rank_order_corr_multi_combine, adjusted_r2_multi_combine = run_nmf_pca(df, target_column, model_name)
        results.append({
            'Dataset': name,
            'Method': 'PCA+NMF-Multiple Combo',
            'Best_Model': best_model_multi_combine,
            'Best_Params': best_params_multi_combine,
            'Best_Score': best_score_multi_combine,
            'MSE': mse_multi_combine,
            'R2': r2_multi_combine,
            'RMSE': rmse_multi_combine,
            'MAE': mae_multi_combine, 
            'Rank-Order Corr': rank_order_corr_multi_combine,
            'Adjusted R': adjusted_r2_multi_combine
        })

    results_df = pd.DataFrame(results)

    results_df.to_csv('results_summary.csv', index=False)
    
    return results_df
```

```python
from sklearn.decomposition import PCA

def apply_pca(data, n_components=3):
    pca = PCA(n_components=n_components)
    transformed_data = pca.fit_transform(data)
    return transformed_data, pca
```

```python
from sklearn.decomposition import NMF

def apply_nmf(data, n_components=5):
    nmf = NMF(n_components=n_components, init='random', random_state=0, max_iter=500)
    features = nmf.fit_transform(data)
    return features, nmf
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

def rmse_score(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse_score, greater_is_better=False)

def grid_search(X_train, y_train):
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('model', LinearRegression())
    ])
    
    param_grid = {
        'model__fit_intercept': [True, False]
    }
    
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring=rmse_scorer)
    grid_search.fit(X_train, y_train)
    
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

def rmse_score(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse_score, greater_is_better=False)

def grid_search_all(X_train, y_train, model_type='LR'):
    print('model type = ', model_type)
    if model_type == 'LR':
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', LinearRegression())
        ])
        param_grid = {
            'model__fit_intercept': [True, False]
        }
    
    elif model_type == 'SVR':
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', SVR())
        ])
        param_grid = {
            'model__C': [0.1, 1.0, 10.0],
            'model__kernel': ['linear', 'rbf']
        }
    
    elif model_type == 'RF':
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', RandomForestRegressor())
        ])
        param_grid = {
        'model__n_estimators': [50, 100, 200, 300, 500],
        'model__max_features': ['auto', 'sqrt', 'log2'],
        'model__max_depth': [None, 10, 20, 30, 40, 50]
    }
    
    elif model_type == 'XGBOOST':
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', XGBRegressor())
        ])
        param_grid = {
            'model__n_estimators': [50, 100, 200],
            'model__learning_rate': [0.01, 0.1, 0.2]
        }
    
    else:
        raise ValueError("Unsupported model_type. Choose from 'LR', 'SVM', 'RF', 'XGBOOST'.")

    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring=rmse_scorer)
    grid_search.fit(X_train, y_train)
    
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, make_scorer
from scipy.stats import spearmanr
import numpy as np

def adjusted_r2_score(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)


def predict_and_evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, predictions)
    rank_order_corr, _ = spearmanr(y_test, predictions)
    
    n = len(y_test)
    k = X_test.shape[1]
    adjusted_r2 = adjusted_r2_score(r2, n, k)
    
    return mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
def handle_negative_values(data):
    min_value = data.min().min()
    
    if min_value < 0:
        data += abs(min_value)
    return data
```

```python
from sklearn.pipeline import Pipeline

def complete_workflow_pca(dataframe, target_column='UPDRS1', model='LR'):

    target_df, dataframe = extract_target_columns(dataframe)
    
    best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2 = run_pca(dataframe, target_df[target_column], model)
    
    return best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
def run_pca(dataframe, target_column, model_name):
    X_train, X_test, y_train, y_test = split_data(dataframe, target_column)
    
    best_score = float('inf')  
    best_params = None
    best_model = None
    best_pca_component = None
    
    
    for n_components in range(2, 6):
        pipeline = Pipeline([
            ('normalizer', MinMaxScaler()),  
            ('pca', PCA(n_components=3))
        ])

        X_train_transformed = pipeline.fit_transform(X_train)
        
        model, params, score = grid_search_all(X_train_transformed, y_train, model_name)
        
        if score < best_score:
                best_score = score
                best_params = params
                best_model = model
                best_pca_component = n_components
    
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    pca = PCA(n_components=best_pca_component)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    best_model.fit(X_train_pca, y_train)
    
    mse, r2, rmse, mae, rank_order_corr, adjusted_r2 = predict_and_evaluate(best_model, X_test_pca, y_test)
    
    print(f"{'Metric':<20} {'Value'}")
    print(f"{'-'*30}")
    print(f"{'Best PCA Component':<20} {best_pca_component}")
    print(f"{'Best Model Parameters':<20} {best_params}")
    print(f"{'Best Score':<20} {best_score}")
    print(f"{'MSE on Test Data':<20} {mse}")
    print(f"{'R on Test Data':<20} {r2}")

    
    
    return best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
def run_nmf(dataframe, target_column, model_name='LR'):
    X_train, X_test, y_train, y_test = split_data(dataframe, target_column)
    
    X_train = handle_negative_values(X_train)
    X_test = handle_negative_values(X_test)
    
    best_score = float('inf')  
    best_params = None
    best_model = None
    best_nmf_component = None
    
    for n_components in range(2, 7):
        nmf = NMF(n_components=n_components, random_state=42, max_iter=1000)
        X_train_nmf = nmf.fit_transform(X_train)
        
        model, params, score = grid_search_all(X_train_nmf, y_train, model_name)
        
        if score < best_score:
            best_score = score
            best_params = params
            best_model = model
            best_nmf_component = n_components
    
    nmf = NMF(n_components=best_nmf_component, random_state=42, max_iter=1000)
    X_train_nmf = nmf.fit_transform(X_train)
    X_test_nmf = nmf.transform(X_test)
    
    best_model.fit(X_train_nmf, y_train)
    
    mse, r2, rmse, mae, rank_order_corr, adjusted_r2 = predict_and_evaluate(best_model, X_test_nmf, y_test)
    
    print(f"{'Metric':<20} {'Value'}")
    print(f"{'-'*30}")
    print(f"{'Best NMF Component':<20} {best_nmf_component}")
    print(f"{'Best Model Parameters':<20} {best_params}")
    print(f"{'Best Score':<20} {best_score}")
    print(f"{'MSE on Test Data':<20} {mse}")
    print(f"{'R on Test Data':<20} {r2}")
    
    return best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
def complete_workflow_multiple_combination(dataframe, target_column='UPDRS1', model_name='LR'):
    target_df, dataframe = extract_target_columns(dataframe)
    
    best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2 =  run_nmf_pca(dataframe, target_df[target_column], model_name)
    
    return best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
def run_nmf_pca(dataframe,target_column, model_name='LR'):
    X_train, X_test, y_train, y_test = split_data(dataframe, target_column)
    
    best_score = float('inf')  
    best_params = None
    best_model = None
    
    
    for pca_components in [2]:
        for nmf_components in [2, 3, 4, 5, 6]:
       
            pca_pipeline = Pipeline([
                ('normalizer', MinMaxScaler()),  
                ('pca', PCA(n_components=pca_components))  
            ])

            X_train_pca = pca_pipeline.fit_transform(X_train)
            
            nmf_pipeline = Pipeline([
                ('nmf', NMF(n_components=nmf_components, random_state=42, max_iter=1000))  
            ])

            X_train_nmf = nmf_pipeline.fit_transform(X_train)
            
            X_train_combined = np.hstack((X_train_pca, X_train_nmf))

            model, params, score = grid_search_all(X_train_combined, y_train, model_name)

            if score < best_score:
                best_score = score
                best_params = {'pca_components': pca_components, 'nmf_components': nmf_components}
                best_model = model
    
    # Evaluate the best model on the test set using the best parameters
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    pca = PCA(n_components=best_params['pca_components'])
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    nmf = NMF(n_components=best_params['nmf_components'], random_state=42, max_iter=1000)
    X_train_nmf = nmf.fit_transform(X_train)
    X_test_nmf = nmf.transform(X_test)
    
    X_train_combined = np.hstack((X_train_pca, X_train_nmf))
    X_test_combined = np.hstack((X_test_pca, X_test_nmf))
    
    best_model.fit(X_train_combined, y_train)
    
    mse, r2, rmse, mae, rank_order_corr, adjusted_r2 = predict_and_evaluate(best_model, X_test_combined, y_test)
    
    print(f"{'Metric':<20} {'Value'}")
    print(f"{'-'*30}")
    print(f"{'Best PCA Component':<20} {best_params['pca_components']}")
    print(f"{'Best NMF Component':<20} {best_params['nmf_components']}")
    print(f"{'Best Model Parameters':<20} {best_params}")
    print(f"{'Best Score':<20} {best_score}")
    print(f"{'MSE on Test Data':<20} {mse}")
    print(f"{'R on Test Data':<20} {r2}")

    
    return best_model, best_params, best_score, mse, r2, rmse, mae, rank_order_corr, adjusted_r2
```

```python
import numpy as np
from sklearn.decomposition import NMF, PCA, FastICA

X = bl_v04_v06_v08_v10_v12_dataframe # Replace this with other dataframe

# NMF
nmf = NMF(n_components=5, init='random', random_state=0, max_iter=500)
X_nmf = nmf.fit_transform(X)

# PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)

# ICA
ica = FastICA(n_components=5, random_state=0)
X_ica = ica.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# K-means clustering
kmeans_nmf = KMeans(n_clusters=3, random_state=0).fit(X_nmf)
kmeans_pca = KMeans(n_clusters=3, random_state=0).fit(X_pca)
kmeans_ica = KMeans(n_clusters=3, random_state=0).fit(X_ica)

# GMM clustering
gmm_nmf = GaussianMixture(n_components=3, random_state=0).fit(X_nmf)
gmm_pca = GaussianMixture(n_components=3, random_state=0).fit(X_pca)
gmm_ica = GaussianMixture(n_components=3, random_state=0).fit(X_ica)
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def plot_clusters(ax, X, labels, model_name):
    unique_labels = set(labels)
    colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels)))
    for i, (label, color) in enumerate(zip(unique_labels, colors)):
        ax.scatter(X[labels == label, 0], X[labels == label, 1], c=[color], label=f'Cluster {label}')
    ax.set_title(f'{model_name} clustering')
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.legend()

fig, axes = plt.subplots(3, 2, figsize=(15, 15))

plot_clusters(axes[0, 0], X_nmf, kmeans_nmf.labels_, 'K-means on NMF')
plot_clusters(axes[0, 1], X_nmf, gmm_nmf.predict(X_nmf), 'GMM on NMF')
plot_clusters(axes[1, 0], X_pca, kmeans_pca.labels_, 'K-means on PCA')
plot_clusters(axes[1, 1], X_pca, gmm_pca.predict(X_pca), 'GMM on PCA')
plot_clusters(axes[2, 0], X_ica, kmeans_ica.labels_, 'K-means on ICA')
plot_clusters(axes[2, 1], X_ica, gmm_ica.predict(X_ica), 'GMM on ICA')

plt.tight_layout()
plt.show()
```
